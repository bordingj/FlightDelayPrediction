# Case 1: Predicting departure delays for an airliner

## Summary
The goal of the present project was to predict the departure delays from JFK airport. We were provided with features; Distance, Destination, Tail-number, Time of departure and Date. Calculated features; Time between departures and Number of departures per day were added and Daily weather information was gathered as well. The generalization error was estimated using 10-fold Cross Validation (CV). Simple linear regression resulted in a RMSE of 31.916. Adding weather data decreased the linear regression RMSE to 31.601. Using the python library Scikit-learn we trained Random Forest (RF) and Gradient Boosting Regressor (GB) models. Moreover, we implemented a 1-hidden layer Neural Network (NN) on the GPU using the python library Theano and made it compatible with the Scikit-learn Bagging meta-algorithm resulting in a Bagged Neural Network model (Bagged NN). Optimization of the hyperparameters of the 3 models was done on 3 different sets of random CV splits (in hope of diversifying the models) and performance was tested on a fourth set of splits with the following RMSE results: RF, 27.773; GB, 28.385; Bagged NN, 28.83. Hence, RF had the best RMSE but the performance of the 3 models were quit similar. As a final model we combined the 3 models to an emsemble by weighted averaging. The final ensemble had a CV RMSE of 27.765. Finally, the ensemble was trained on the whole data set and predictions for the evalutation set was submitted.

We want to construct a baseline for our estimator such that we can evaluate whether we are improving or not.

Therefore we randomly shuffle the data and generate 10 different sets of indices that will split the data into a training and validation set using the statistical learning library Scikit-learn.


As baseline estimator we just use simple __multivariate linear regression__ with Tail-number and Destination encoded as dummy variables. Tail-numbers/Destinations with less than 20 occurrences were excluded.

```python
kf = prob.GenerateCVindexes(n_folds=10)
from sklearn.linear_model import LinearRegression
RMSE = prob.CrossValidation(Regr=LinearRegression(), kf=kf)
print('RMSE: {}'.format(RMSE))
```

## Random Forest

As a first model we choose __Random Forest regressor__ from __Scikit Learn__. The Random Forest regressor are bagged (boostrap aggregated) regression trees where the trees are grown such that at each candidate split only a random subset of the features are consideres ('feature bagging'). If we have $M$ different trees then our prediction is
\begin{align}
f(\mathbf{x})= \sum_{m=1}^M \frac{1}{M} f_m(\mathbf{x})
\end{align}
where $f_m$ is the $m$'th tree.

### Selecting Hyperparameters

One parameter of the random forest is the __minimum number of samples required to split an internal node__ (which in turn control the size of each tree). In order to choose this parameter and the overall __number of features__ to use we perform a CV grid search and choose the minimum RMSE. Random forest does not overfit with increasing number of tree. For choosing the parameters we set the number of trees to 200.

```python
feature_importance_CV_list_RF = pickle.load(open('feature_importance_CV_list_RF.p', 'rb'))
kf_RF = pickle.load(open('kf_RF.p', 'rb'))
from sklearn.ensemble import RandomForestRegressor

no_estimators = 200 #number of trees
min_samples_split_linspace = np.linspace(10,40,7,dtype=int)
no_features_linspace = np.linspace(10,40,7,dtype=int)
##### CV GRID SEARCH FOR OPTIMAL NUMBER OF FEATURES AND MIN-SAMPLES-SPLIT
CV_GRID_RF = np.zeros((min_samples_split_linspace.shape[0],
                       no_features_linspace.shape[0]))
for i, min_samples_split in enumerate(min_samples_split_linspace):
    for j, no_features in enumerate(no_features_linspace):
        Regr = RandomForestRegressor(n_estimators=no_estimators, 
                                    min_samples_split=min_samples_split, 
                                    n_jobs=4)
        RMSE = prob.CrossValidation(Regr=Regr, kf=kf_RF, use_y_log = True, 
                                    standardization=True, 
                                    feature_importance_list=feature_importance_CV_list_RF,
                                    num_features=no_features)
        CV_GRID_RF[i,j] = RMSE
#Save results
np.save(file='CV_GRID_RF',arr=CV_GRID_RF)
#plot
(best_RMSE, opt_split, opt_features) = plot_CV_GRID_and_getOptimum(CV_GRID_RF, 
                                                        min_samples_split_linspace,
                                                        no_features_linspace, 
                                                        'min_samples_split',
                                                        'no_features',
                                                        'CV GRID for Random Forest Regreessor')
print('Best RMSE of {0:0.3f} at a min_samples_split of {1} and {2} features'.format(
                                            best_RMSE, opt_split, opt_features))
#Save model
Regr = RandomForestRegressor(n_estimators=500, min_samples_split=opt_split, n_jobs=4)
joblib.dump(Regr, 'RF.pkl')
```

We get an optimal __20 features__ and optimal __min_samples_split of 20__ at a minimum CV RMSE of 27.799. For the final random forest model we increase the number of trees to 500. 

## Gradient Boosting Regressor
As a second model we choose the __Gradient Boosting Regressor__ from __Scikit-learn__. __Boosting__ is an algorithm that apply base learners sequantially to weighted versions of the data, where more weight is given to examples with larger residuals (for regression) or those which are misclassified (classification). In the Gradient Boosting regressor in each stage a regression tree is fit on the negative gradient of the given loss function. We choose the least squares loss function.

### Selecting Hyperparameters

For the gradient boosting regressor, the __number of trees__ is a parameter of the model. Too many trees may overfit the data. Moreover, the __learning rate__ is also a parameter (lower learning rate generally requires more trees). We do a CV grid search on order to choose the learning rate and the number of trees. The min_sample_split and the overall number of features was set to the same as for Random Forest.

```python
feature_importance_CV_list_GB = pickle.load(open('feature_importance_CV_list_GB.p', 'rb'))
kf_GB = pickle.load(open('kf_GB.p', 'rb'))

from sklearn.ensemble import GradientBoostingRegressor

learning_rate_linspace = np.linspace(0.1,0.5,5)
no_estimators_linspace = np.linspace(800,1200,5,dtype=int)
##### CV GRID SEARCH FOR OPTIMAL NUMBER OF FEATURES AND MIN-SAMPLES-SPLIT
CV_GRID_GB = np.zeros((learning_rate_linspace.shape[0],no_estimators_linspace.shape[0]))
# Run in parallel
import multiprocessing
from functools import partial

def full_func(learning_rate, no_estimators):
    Regr = GradientBoostingRegressor(n_estimators=no_estimators, 
                                    min_samples_split=opt_split,
                                    learning_rate=learning_rate)
    RMSE = prob.CrossValidation(Regr=Regr, kf=kf_GB, use_y_log = True, 
                                standardization=True, 
                                feature_importance_list=feature_importance_CV_list_GB,
                                num_features=opt_features)
    return RMSE
pool=multiprocessing.Pool() 
for j, no_estimators in enumerate(no_estimators_linspace):
     partial_func=partial(full_func,no_estimators=no_estimators)     
     CV_GRID_GB[:,j]=pool.map(partial_func,learning_rate_linspace)
pool.close()
#Save results
np.save(file='CV_GRID_GB',arr=CV_GRID_GB)

#plot
(best_RMSE, opt_learning_rate, opt_no_estimators) = plot_CV_GRID_and_getOptimum(
                                                        CV_GRID_GB, 
                                                        learning_rate_linspace ,
                                                        no_estimators_linspace, 
                                                        'learning rate',
                                                        'no_estimators',
                                                        'CV GRID for Gradient Boosting Regreessor')
print('Best RMSE of {0:0.3f} at with {1} trees at learning rate {2:0.1f}'.format(
                                                                best_RMSE, 
                                                                opt_no_estimators, 
                                                                opt_learning_rate))
#Save model
Regr = GradientBoostingRegressor(n_estimators=opt_no_estimators, 
                                    min_samples_split=opt_split,
                                    learning_rate=opt_learning_rate)
joblib.dump(Regr, 'GB.pkl')
```

We get a minimum RMSE of 28.320 with __1000 trees__ at a __learning rate of 0.3__.

## Bagged Neural Nets

For a third model we construct __Bagged Neural Networks__. Training alot of neural networks is not scalable on the CPU, so using the python GPU library __Theano__ we implemented a simple Neural Network with __one hidden layer__ with a hyperbolic tangent as basis function from scratch which is compatible with the Scikit-Learn. We train the Neural Network on a GTX Geforce 970. 

### Selecting Hyperparameters for simple Neural Network
We do a CV grid search for choosing __the number of hidden units__ and the __regularization constant__. Using not the logtransformed dependent variable worked better for the neural network

```python
feature_importance_CV_list_BaggedNN = pickle.load(
                                            open('feature_importance_CV_list_BaggedNN.p', 'rb'))
kf_BaggedNN = pickle.load(open('kf_BaggedNN.p', 'rb'))

from NN import MLP_Regressor
L2_reg_linspace = np.linspace(1, 3, 5)
no_units_linspace = np.linspace(30, 120, 10,dtype=int)
CV_GRID_NN = np.zeros((L2_reg_linspace.shape[0],no_units_linspace.shape[0]))
for i, L2_reg in enumerate(L2_reg_linspace):
    for j, num_hidden_units in enumerate(no_units_linspace):
        Regr = MLP_Regressor(num_hidden_units, L2_reg)
        RMSE = prob.CrossValidation(Regr=Regr, kf=kf_BaggedNN, use_y_log = False, 
                                    standardization=True, 
                                    feature_importance_list=feature_importance_CV_list_BaggedNN,
                                    num_features=opt_features)
        CV_GRID_NN[i,j] = RMSE
#Save results
np.save(file='CV_GRID_NN',arr=CV_GRID_NN)

#plot
(best_RMSE, opt_L2_reg, opt_no_units) = plot_CV_GRID_and_getOptimum(CV_GRID_NN, 
                                                                    L2_reg_linspace,
                                                                    no_units_linspace, 
                                                                    'L2_reg',
                                                                    'No. hidden units',
                                                                    'CV GRID for Neural Network')
print('Best RMSE of {0:0.3f} at with {1} hidden units with regularization constant of {2:0.1f}'.format(
                                                                                best_RMSE, 
                                                                                 opt_no_units, 
                                                                                 opt_L2_reg))
#Save model
Regr = MLP_Regressor(opt_no_units, opt_L2_reg)
joblib.dump(Regr, 'NN.pkl')
```

We get a RMSE of 29.028 with __120 hidden units__ with __regularization constant of 2.0__

### Building Bagged Neural Nets

We feed the simple neural network to the __Scikit Learn__ meta-algorithm __Bagging Regressor__, set the number of estimators to 100 and save the model.

```python
from sklearn.ensemble import BaggingRegressor
Base_Regr = joblib.load('NN.pkl')
Regr = BaggingRegressor(base_estimator=Base_Regr, n_estimators=100)
joblib.dump(Regr, 'Bagged_NN.pkl')
```

## Estimating the RMSE of the 3 models

For each of the final models we estimate the RMSE by doing 10-fold CV.

```python
RF = joblib.load('RF.pkl')
GB = joblib.load('GB.pkl')
Bagged_NN = joblib.load('Bagged_NN.pkl')

feature_importance_CV_list_ensemble = pickle.load(
                                        open('feature_importance_CV_list_ensemble.p', 'rb'))
kf_ensemble = pickle.load(open('kf_ensemble.p', 'rb'))

models = [RF, GB, Bagged_NN]
RMSEs = []
for Regr in models:
    if Regr == Bagged_NN:
        use_y_log = False
    else:
        use_y_log = True
    RMSE = prob.CrossValidation(Regr=Regr, kf=kf_ensemble, use_y_log = use_y_log, 
                                standardization=True, 
                                feature_importance_list=feature_importance_CV_list_ensemble,
                                num_features=opt_features)
    RMSEs.append(RMSE)
model_names = ['RF', 'GB', 'Bagged_NN']
models_and_RMSEs = dict(zip(model_names,zip(models, RMSEs)))
print(dict(zip(model_names, RMSEs)))
pickle.dump(models_and_RMSEs, open('models_and_RMSEs.pkl', 'wb'))
```

We get RMSEs of:
   - Random Forest: 27.773
   - Gradient Boosting: 28.385
   - Bagged Neural Networks: 28.83

## Building Ensemble of the 3 models

We wish to build an ensemble of the 3 models. We combine the models based on their CV RMSE as follows:

\begin{equation}
f(\mathbf{x}) = \sum_{m=1}^3 w_m f_m(\mathbf{x})
\end{equation}
where \begin{equation}w_m = \frac{(RMSE_m-c)^{-1}}{\sum_{m'=1}^3 (RMSE_{m'}-c)^{-1}}
\hspace{0.5cm}
\text{ and } \hspace{0.5cm}
c = \min\{RMSE_{RF}, RMSE_{BG}, RMSE_{Bagged\_{NN}}\}-1
\end{equation}

```python
models_and_RMSE = pickle.load(open('models_and_RMSEs.pkl', 'rb'))

Ensemble = MyEnsemble(RF=models_and_RMSE['RF'][0], RF_RMSE=models_and_RMSE['RF'][1], 
                      GB=models_and_RMSE['GB'][0], GB_RMSE=models_and_RMSE['GB'][1], 
                      Bagged_NN=models_and_RMSE['Bagged_NN'][0], 
                      Bagged_NN_RMSE=models_and_RMSE['Bagged_NN'][1])
```

## Estimating RMSE of Ensemble
We do a final estimate of the RMSE

```python
feature_importance_CV_list_ensemble = pickle.load(
                                            open('feature_importance_CV_list_ensemble.p', 'rb'))
kf_ensemble = pickle.load(open('kf_ensemble.p', 'rb'))

RMSE = prob.EstimateRMSEforMyEnsemble(Regr=Ensemble, kf=kf_ensemble, standardization=True, 
                                      feature_importance_list=feature_importance_CV_list_ensemble, 
                                      num_features=opt_features)
print('RMSE of Ensemble: {0: 0.3f}'.format(RMSE))
f = open('RMSE of Ensemble.txt', 'w')
f.write('RMSE of Ensemble: {}'.format(RMSE))
f.close()
```

We get the lowest yet __RMSE of 27.765__ and we should have good confidence in this estimate since the optimization of the model parameters have been done on CV splits different from the one for estimating this estimate.
